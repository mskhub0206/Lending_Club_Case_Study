**Check the null values**

```python
a = loan_data.isnull().sum()
b = len(loan_data)
pt.figure(figsize=(20,4))
((a)/b*100).plot.bar(title='Check the null values', color='blue')
```

####  The above picture shows the missing values in the data set

#### Removing the outliers is very important as many datas are missed out, based on the above plot the following code remove the missing values.

```python
# the code iterates through the missing columns that have the missing values.
missing_columns = loan_data.columns[loan_data.isna().any()]
for column in missing_columns:
    missing_values = loan_data[column].isna().sum()
    portion = (missing_values / loan_data.shape[0]) * 100
    print(f"'{column}': missing values '{missing_values}' ==> '{portion:.1f}%'")
```

```python
# This code creates a bar graph using the seaborn library. The x-axis is with the "Purpose Titles" and the y-axis is labeled "Count". 
# Additionally, the code adds percentages to each bar to indicate what percentage of loans were taken out for each purpose. 
pt.figure(figsize=(14,6))
newdata = sb.countplot(x='purpose', data=loan_data, color='blue')
newdata.set_title("Client Purposes for Loan Credit", fontsize=22)
newdata.set_xlabel("Purpose Titles", fontsize=18)
newdata.set_ylabel('Count', fontsize=18)
sizes = [value.get_height() for value in newdata.patches]
newdata.set_ylim(0, max(sizes) * 1.10)
for value in newdata.patches: 
    height = value.get_height() 
    newdata.text(value.get_x()+value.get_width()/2., height + 3, '{:1.2f}%'.format(height/total*100), ha="center", fontsize=14) 
newdata.set_xticklabels(newdata.get_xticklabels(), rotation=45) 
pt.show() 
```

**Conclusion:**

1. Based on the above graph the debt_consolidation purpose contributesapproximately 47% of overall loan borrowers.
2. Also, credit card users contributes to almost 13% of loan borrowers.

```python
loan_purpose = ['purpose', 'loan_status']
normalized_table = round(pd.crosstab(loan_data[loan_purpose[0]], loan_data[loan_purpose[1]], normalize='columns') * 100, 2)
style = sb.light_palette("green", as_cmap=True)
normalized_table.style.background_gradient(cmap=style)
```
